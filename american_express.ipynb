```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("data/train_data.csv")
train_label = pd.read_csv("data/train_labels.csv")
```

```python
def handleD63(data):
    value_names = np.unique(data.D_63)
    df = pd.DataFrame()
    for i, val in enumerate(value_names):
        df[f"D_63_{val}"] = 1.0 * (data.D_63 == val)
    return df

def handleD64(data):
    value_names = np.unique(data.D_64.astype(str))
    df = pd.DataFrame()
    for i, val in enumerate(value_names):
        if val == 'nan' or val == '-1': continue
        df[f"D_64_{val}"] = 1.0 * (data.D_64 == val)
    return df
```

```python
train_data = pd.concat([train_data, handleD63(train_data), handleD64(train_data)], axis = 1)
train_data.drop(columns = ["D_63", "D_64"], inplace=True)
```

```python
# train_data.head()
# S_2 is date
train_data["S_2"] = pd.to_datetime(train_data.S_2)
```

```python
train_data.S_2.values.astype(np.float64)
```

```python
train_data.S_2 = train_data.S_2.values.astype(np.float64)
```

```python
train_data.B_31 = train_data.B_31.astype(np.float64)
```

```python
train_label = pd.read_csv("data/train_labels.csv")
```

```python
train_data_groupby_id = train_data.groupby("customer_ID").agg('mean')
train_data_groupby_id
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
```

```python
train_data_groupby_id_merge_label = train_data_groupby_id.merge(train_label, on="customer_ID")
train_data_groupby_id_merge_label.head()
```

```python
y = train_data_groupby_id_merge_label.target
y.head()
```

```python
train_data_groupby_id_merge_label.drop(columns=["target"], inplace=True)
train_data_groupby_id_merge_label.head()
```

```python
train_data_groupby_id_merge_label.drop(columns=['customer_ID'], inplace=True)
train_data_groupby_id_merge_label.head()
```

```python
X = train_data_groupby_id_merge_label
X.head()
```

```python
nan_columns = X.columns[X.isna().any()].tolist()
```

```python
X.drop(columns=nan_columns, inplace=True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
clf.fit(X_train, y_train)
```

```python
prediction = clf.predict(X_test)
```

```python
np.sum(np.abs(y_test - prediction)) / len(y_test)
```
